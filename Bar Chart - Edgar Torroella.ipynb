```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = df_used_cars["Brand"], 
        height = df_used_cars["Cars Listings"],
        color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold", color="blue")
plt.ylabel("Number of Listings", fontsize = 13 )
plt.savefig("Cars Bars.png")
plt.show()
```
